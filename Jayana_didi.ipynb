```python
#use the Python 3.8.18 environment on anaconda
#Creation of dataframes from the .csv file
%pip install tabulate
import pandas as pd
import csv

file_path = "D:/Nuclear Project/21_08_25/gamma pulses_1.csv"
parsed_data = []
header = []
with open(file_path, 'r') as f:
    csv_reader = csv.reader(f, delimiter=';')
 
    # Read the header
    header = next(csv_reader)
 
    # Read the data rows
    for row in csv_reader:
        if not row:
            continue

        # The first 6 columns are fixed, the rest are part of SAMPLES
        fixed_cols = row[:6]
        samples = row[6:]
 
        # Create a dictionary for the row, joining samples into a list
        parsed_row = {

            'BOARD': fixed_cols[0],

            'CHANNEL': fixed_cols[1],

            'TIMETAG': fixed_cols[2],

            'ENERGY': fixed_cols[3],

            'ENERGYSHORT': fixed_cols[4],

            'FLAGS': fixed_cols[5],

            'SAMPLES': samples

        }
        parsed_data.append(parsed_row)
 
# Create the DataFrame

df = pd.DataFrame(parsed_data)
print("First 5 rows of the DataFrame:")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

```

```python
# ...existing code...

# Expand SAMPLES into separate columns t0, t1, ...
samples_df = pd.DataFrame(df['SAMPLES'].tolist())
samples_df = samples_df.add_prefix('t')

# Concatenate the new columns with the original DataFrame (excluding the old SAMPLES column)
df_expanded = pd.concat([df.drop(columns=['SAMPLES']), samples_df], axis=1)

# print("First 5 rows with expanded SAMPLES columns:")
# print(df_expanded.head().to_markdown(index=False, numalign="left", stralign="left"))
# ...existing code...
df_expanded.head()
```

```python
#inserting the data in the MySQL database
import pandas as pd
 
# Assuming `df` is your DataFrame and `cursor` and `connection` are set up
 
# Dynamically create column names t0 to t995
t_columns = [f"t{i}" for i in range(996)]
 
# Full list of columns including initial ones
columns = ["BOARD","CHANNEL","TIMETAG","ENERGY","ENERGYSHORT","FLAGS"] + t_columns
 
# Create the insert query dynamically in a single line
insert_query = f"INSERT INTO `your_table_name` ({', '.join(columns)}) VALUES ({', '.join(['%s']*len(columns))});"
 
# Replace placeholder table name
table_name = "gamma"
insert_table_query = insert_query.replace("your_table_name", table_name)
 
# Prepare data from DataFrame (make sure your df has all required columns)
data = [tuple(row[col] for col in columns) for index, row in df_expanded.iterrows()]

# ...existing code...

batch_size = 50  # Try 100, or lower if you still get errors

try:
    cursor = connection.cursor()
    for start in range(0, len(data), batch_size):
        end = start + batch_size
        batch = data[start:end]
        cursor.executemany(insert_table_query, batch)
        connection.commit()
        print(f"Inserted rows {start} to {end-1}")
    print(f"‚úÖ Data successfully inserted into table `{table_name}`.")
except Exception as e:
    print(f"‚ùå Failed to insert data into table `{table_name}`. Error: {e}")
```
